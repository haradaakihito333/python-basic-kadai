```python
class Human:
  #コンストラクタの定義
  def __init__(self,name,age):
    self.name = name
    self.age = age

  #メソッドの定義
  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name}は大人です。")
    else:
      print(f"{self.name}は大人ではありません。")

#インスタンス化（リスト
human_rist = {"原田明人":51, "原田亜由美":51, "原田はるき":14}
for name, age in human_rist.items():
  index = Human(name,age)
  #メソッドにアクセス
  index.check_adult()
```
